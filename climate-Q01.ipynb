```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

from pathlib import Path

# Small style adjustments for more readable plots
plt.style.use("seaborn-v0_8-whitegrid")
plt.rcParams["figure.figsize"] = (8, 6)
plt.rcParams["font.size"] = 14
```

```python
DATA_DIR = Path.home()/Path('shared/climate-data')

co2_file = DATA_DIR / "monthly_in_situ_co2_mlo_cleaned.csv"
data = pd.read_csv(co2_file, na_values=-99.99).dropna()
```

```python
fig = sns.lineplot(data=data, x='month', y='c02', hue='year')
sns.move_legend(fig, "upper left", bbox_to_anchor=(1, 1))
fig.get_figure().savefig("outputs/Q01.png", bbox_inches="tight")
```
